```python
import torch
from diffusers import StableDiffusionPipeline
import gradio as gr
```

```python
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5"
)

pipe = pipe.to("cuda")  # use "cuda" if GPU available

print("✅ Model Loaded")
```

```python
def generate_face(age, smile, hair, gender):

    prompt = f"""
    realistic human face, {gender}, age {age},
    {'smiling' if smile=='yes' else 'serious'},
    {hair} hair, high quality portrait
    """

    image = pipe(
        prompt,
        num_inference_steps=20   # 🔥 reduce from 50 → 20
    ).images[0]

    return image
```

```python
iface = gr.Interface(
    fn=generate_face,
    inputs=[
        gr.Slider(0, 80, value=25, label="Age"),
        gr.Radio(["yes", "no"], value="yes", label="Smile"),
        gr.Dropdown(["black", "brown", "blonde"], value="black", label="Hair Color"),
        gr.Radio(["male", "female"], value="male", label="Gender")
    ],
    outputs=gr.Image(height=400, width=400),

    title="🧠 Face Generator using DCGAN",

    description="""
    This system generates realistic human face images using Deep Convolutional
    Generative Adversarial Networks (DCGAN). The generator model learns to
    produce synthetic facial images from random noise, while the discriminator
    evaluates the authenticity of generated images. The system allows users
    to control attributes such as age, gender, smile, and hair characteristics.
    """,

    theme="soft"
)

iface.launch(share=True)
```
